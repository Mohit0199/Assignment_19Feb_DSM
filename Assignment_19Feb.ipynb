```python
# Q3. What is App routing in Flask? Why do we use app routes?

## In Flask, app routing refers to the process of mapping URLs to Python functions within a Flask application.
## It allows to define different URL endpoints (routes) and specify the Python functions that should be executed
## when those endpoints are accessed by clients (e.g., web browsers).

## app routing in Flask is essential for defining the URLs and mapping them to corresponding Python functions,
## allowing for dynamic request handling and creating modular, flexible web applications.
```
